# Map old user_id into new user_id

```python
ad_events_df = ad_events_df.merge(
    users_df[['user_id', 'new_user_id']], # get 2 columns user_id and new_user_id from users_df
    on = 'user_id', # join by user_id
    how = 'left' # left join - keep all rows from ad_events_df
)
```

```python
# Delete user_id column
ad_events_df = ad_events_df.drop('user_id', axis=1)

# Rename new_user_id to user_id
ad_events_df = ad_events_df.rename(columns={'new_user_id': 'user_id'})
```

```python
# Change columns order 
cols = ['event_id', 'ad_id', 'user_id', 'timestamp', 'day_of_week', 'time_of_day', 
        'event_type', 'date', 'time', 'day', 'month', 'quarter', 'year', 
        'hour', 'minute', 'second']
ad_events_df = ad_events_df[cols]
```

```python
ad_events_df.sample(3)
```

# Delete user_id column & change new_user_id in users_df

```python
# Delete user_id column
users_df = users_df.drop('user_id', axis=1)

# Rename new_user_id to user_id
users_df = users_df.rename(columns={'new_user_id': 'user_id'})
```

```python
cols = ['user_id', 'user_gender', 'user_age', 'age_group', 'country', 'location', 'interests']
users_df = users_df[cols]
```
